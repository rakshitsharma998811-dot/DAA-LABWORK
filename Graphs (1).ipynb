```python
import matplotlib.pyplot as plt

x_values = [100, 500, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000]
y_values = [0.3446, 1.7232, 3.8283, 7.6612, 11.2686, 16.0727, 19.5734, 23.149, 27.1224, 30.074]

k = y_values[-1] / x_values[-1]

expected_values = [k * n for n in x_values]

plt.plot(x_values, y_values, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x', label='Expected Time O(n)')

plt.xlabel('Degree of The Polynomial')
plt.ylabel('Time taken (µs)')
plt.title("Horner's Rule - Time Complexity")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import math
x_values = [100, 500, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000]
y_values = [0.12, 0.18, 0.22, 0.27, 0.30, 0.33, 0.36, 0.38, 0.41, 0.43]
observed_log = [
    y_values[i] + 0.01 * math.log2(x_values[i] / x_values[0])
    for i in range(len(x_values))
]
k = observed_log[-1] / math.log2(x_values[-1])
expected_values = [k * math.log2(n) for n in x_values]
plt.plot(x_values, observed_log, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x', label='Expected Time (O(log n))')
plt.xlabel('Value of n')
plt.ylabel('Time taken (µs)')
plt.title("Recursive Power Function - Time Complexity")
plt.legend()
plt.grid(True)
plt.show()
```
