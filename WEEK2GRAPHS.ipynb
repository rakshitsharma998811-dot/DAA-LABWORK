```python

import matplotlib.pyplot as plt
import math

x_values = [1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000]
y_values = [0.03, 0.045, 0.05, 0.055, 0.058, 0.060, 0.062, 0.064, 0.066, 0.068]

k = 0.005
expected_values = [k * math.log2(n) for n in x_values]

plt.plot(x_values, y_values, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x', label='Expected Time O(log n)')

plt.xlabel("Value of n")
plt.ylabel("Time taken (µs)")
plt.title("Recursive Power Function - Time Complexity")

plt.xticks(x_values)
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = [500, 1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000]
y_values = [325.8, 841.8, 3328.3, 4985.8, 10637.4, 17329.4, 17700.9, 25945, 36059.9]

k = y_values[0] / (x_values[0]**2)

expected_values = [k * (n**2) for n in x_values]

plt.plot(x_values, y_values, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x', label='Expected Time O(n²)')

plt.xlabel("Input Size (n)")
plt.ylabel("Time taken (µs)")
plt.title("Insertion Sort - Time Complexity AVERAGE CASE")

plt.xticks(x_values)
plt.legend()
plt.grid(True)

plt.show()
```
