```python
import matplotlib.pyplot as plt
import math

x_values = [1000, 5000, 10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000]
y_values = [0, 938.8, 1327.3, 4669.6, 5260.7, 8079.5, 9791, 9944.8, 11624.5, 16046.1]

k = y_values[1] / (x_values[1] * math.log2(x_values[1]))

expected_values = [k * n * math.log2(n) for n in x_values]

plt.plot(x_values, y_values, marker='o', label='Observed Time')
plt.plot(x_values, expected_values, linestyle='--', marker='x', label='Expected Time O(n log n)')

plt.xlabel("Input Size (n)")
plt.ylabel("Time taken (µs)")
plt.title("Merge Sort - Time Complexity")

plt.xticks(x_values)
plt.legend()
plt.grid(True)

plt.show()
```
